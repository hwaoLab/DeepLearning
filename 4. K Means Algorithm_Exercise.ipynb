```python
import numpy as np

def data_gen(n):
    x=np.random.random(n)
    y=np.random.random(n)
    z=np.ones(n)
    z[(x<0.50) & (y<0.50)]=1
    z[(x>0.50) & (y>0.50)]=4
    z[(x>=0.50) & (y<0.50)]=3
    z[(x<0.50) & (y>=0.50)]=2
    return x,y,z
```

```python
# 거리를 위한 평균 값 구하기 
def class_mean(x,y,z,k):
    u=np.ones(k*2)
    u=u.reshape(k,2)
    
    for i in range(k):
        for j in range(2):
        # 2번 반복에서 0인지, 1인지 check 
           if j==0:u[i,j]=np.array(np.mean(x[z==i+1])) 
           # x[z==i+1] => true, false 반환하여,  x에서 z가 해당 조건인 x만 추출(type이 4개 이니깐 조건 4개)
           # i+1은 index가 0부터 시작하기 때문에,
            
           # 1열이면, x에 대한 평균
           if j==1: u[i,j]=np.array(np.mean(y[z==i+1]))
                
    # u는 z가 제공하는 type에 따라, x,y를 평균 매긴 값을 1,2열에 보관하고 있다. 
            
    return u # 같은 type의 점(xx,yy에 대한 평균 값을 구하여, u값으로 리턴 -> 점 사이의 거리를 구할때, 이용함)

n=10
k=4
x,y,z=data_gen(n)

u=class_mean(x,y,z,k)

def class_k_distance(x,y,uu,k,n):
    xx=x.copy()
    yy=y.copy()
    z=np.ones(n) # 랜덤 생성 크기 만큼 1원소를 행렬에 채움 
    
    # 두 점사이의 길이 구하기(변수따로 선언해서 계산)
    xx1=(xx-uu[0,0])**2+(yy-uu[0,1])**2 # uu[i,0]은 x평균 / u[i,1]은 y평균
    xx2=(xx-uu[1,0])**2+(yy-uu[1,1])**2
    xx3=(xx-uu[2,0])**2+(yy-uu[2,1])**2
    xx4=(xx-uu[3,0])**2+(yy-uu[3,1])**2
    yy=np.array([xx1,xx2,xx3,xx4])
    ww=z.copy() # x,y와 같은 길이의 배열 
    
    for i in range(n):
        dmin=np.argmin(yy[:,i])  
        # yy에서 첫번째 행을 모두 포함한 이유는 [xx1,xx2,xx3,xx4]을 모두 불러와 최소값의 인덱스 찾기위해  => 해당작업을 i번 반복
        # argmin 최소값의 인덱스를 얻고자 할 때 사용
        ww[i]=dmin+1 #인덱스는 0부터 시작하기 때문에 +1  => index
        # ww는 type별 거리에서 최단거리를 파악한후, 해당 index를 보관함. 
             
    return ww
# u 는 x와 y 각각 type 별로 평균 매긴 값을 가진 행렬이다.열은 x -> u[i,0],y -> u[i,1]이기 때문에 2열을 가진다. 


def k_means_algorithm(k,n,x,y,z):
    u=class_mean(x,y,z,k)

    while(1):
        ww=class_k_distance(x,y,u,k,n)
        # x,y 점에 대하여 type별 중심과의 거리를 잰 후, 가장 최소값의 인덱스를 추출한 뒤, +1을 하여 type의 번호를 매겨준다.
        # ww는 현재 xx,yy와 크기가 같으며,어떤 type에 속하는 지 담고 있는 배열이다. 
        uu=class_mean(x,y,ww,k) 
        # 이전에는 3번째 인자에 z(type에 대한 기준 정보)가 들어갔지만, 
        # 이번에는 ww(기준이 적용된 실제type정보) 따라서 값에 nan값이 존재 why? 해당 값만 넣어주었기 때문에, 나머지 nan
    
        # u는 type 기준정보, uu는 기준이 적용된 정보 
        print(uu==u)
        print("sum정보:",np.sum(u==uu))

        if np.sum(u==uu)==k*2:
            # np.sum(array1==array2) 는 array1과 array2간의 같은 원소의 개수를 세주는 역할을 한다.
            # 다시 그룹했을때, 기존과 같이 
            break
        u=uu.copy()
    
    u=uu.copy()
    ww=class_k_distance(x,y,u,k,n)
    return uu, ww
    

def accuracy(z,ww):
    acc1=np.sum(z==ww)/len(z)
    print(acc1)
    return acc1

# main program
import numpy as np
n=10
k=4
x,y,z=data_gen(n)
uu,ww=k_means_algorithm(k,n,x,y,z)
print('uu:',uu)
print('ww:',)
```
